# Employee Satisfaction & Workload Analysis

- Data Cleaning
- Exploring data analysis 
- visualization

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

```

```python
df=pd.read_csv("employee_insights.csv")
df.head(50)
```

```python
df.isnull().sum()
```

```python
df=df.dropna(subset=['hours_worked_weekly'])
df=df.dropna(subset=['overtime_hours'])

df.head(50)
```

```python
df.describe()
```

```python
df=df[['satisfaction_score','hours_worked_weekly']].corr()
df
```

```python
dep_avg=df.groupby('department')['satisfaction_score'].mean().sort_values()
print(dep_avg)
```

```python
plt.figure(figsize=(8,6 ))
sns.scatterplot(x='hours_worked_weekly',y='satisfaction_score',data=df)
sns.regplot(x='hours_worked_weekly', y='satisfaction_score', data=df)
plt.title("working hours VS satisfaction")
plt.show()
```

```python
plt.figure(figsize=(10,6))
dep_avg.plot(kind='bar', color='skyblue')
plt.title("Average Satisfaction by Department")
plt.ylabel("Satisfaction Score")
plt.show()
```

```python
plt.figure(figsize=(8,6))
sns.heatmap(df.corr(), annot=True, cmap='coolwarm', fmt=".2f")
plt.title("Correlation Heatmap", fontsize=14)
plt.xticks(rotation=45)
plt.yticks(rotation=0)
plt.show()
```
