# Promotion and performance insights
- data cleaning 
- check for missing values
- calculate the average :
   - projects
   - hours worked daily 
   - satisfaction score
- Visualization of data set

```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np  

df=pd.read_csv("employee_insights.csv",na_values=[' ','na','Null'])
df.head()
```

```python
df.isnull().sum()
```

```python

df.head()
```

```python
df=df.dropna(subset=['promotion_eligible'])
```

```python
df.head(10)
```

```python
df['promotion_eligible']=df['promotion_eligible'].replace({'Yes':'Y','No':'N'})
df.head(10)
```

```python
df['promotion_eligible']=df['promotion_eligible'].replace({'Y':1,'N':0})
df.head(10)
```

## Finding the performace

```python
avg_perf=df.groupby('promotion_eligible')[['promotion_eligible','hours_worked_weekly','satisfaction_score']].mean()
print(avg_perf)
```

```python
percent_eligible=(df['promotion_eligible'].sum()/len(df))*100
print(f"Percentage of employees eligible for promotion: {percent_eligible:.2f}%")
```

```python
promote_count=df['promotion_eligible'].value_counts()
labels= ['Eligible','Not Eligible']

colors=["#A0AF4C","#AF4C4C"]

plt.Figure(figsize=(6,6))


plt.pie(promote_count, labels=labels, autopct='%1.1f%%', startangle=90, colors=colors)

plt.title('Promotion Eligibility Distribution', fontsize=14)

plt.show()



```
